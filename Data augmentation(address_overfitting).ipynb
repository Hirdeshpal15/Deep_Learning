## Data Augmentation To Address Overfitting In Flower Classification CNN
> will build a CNN to classify flower images.
>
> will check model overfitting and how overfitting can be addressed using data augmentation.
>
> Data augmentation is a process of generating new training samples from current training dataset using transformations such as zoom, rotations, change in contrast etc.
>
> will use tensorflow offical web "https://www.tensorflow.org/tutorials/images/classification" as a reference.
>

## Load flowers dataset

```python
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file(
    fname='flower_photos', 
    origin=dataset_url, 
    cache_dir='.', 
    untar=True
)
```
