# Handwritten digits classification using Neural Network
> will classify handwritten digits using a simple neural network which has only input and output layers.Than will try to add a hidden layer and see if performance of the model improves or not.

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train), (X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
y_train[0]  # on X image and on y = numbers to be predicted
```

```python
# As we have max value is 255 so scaling them in o and 1 so that can provide better result

X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
# convert the from High Dimnetion (3D array) to low dimenation(2D array)

X_train_flattened= X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened.shape
```

```python

X_train_flattened[0]
```

## Neural Network with no hidden layers

```python

model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# giving the accurracy of 92 %
model.evaluate(X_test_flattened, y_test)
```

```python
y_predict = model.predict(X_test_flattened)
```

```python
y_predict[0]
```

#### np.argmax finds a maximum element from an array and returns the index of it


```python
np.argmax(y_predict[0])
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predict]
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,6))
sns.heatmap(cm,annot=True,fmt='.2g')
```

```python
# add fmt = d to reove the e value
import seaborn as sns
plt.figure(figsize = (10,6))
sns.heatmap(cm,annot=True,fmt='d')
```

## Using one hidden layer to improve the accuracy

```python


model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# accuray has been incresed from 92 to 97 %

model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,6))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

### Using Flatten layer so that we don't have to call .reshape on input dataset

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100,activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics= ['accuracy'])

model.fit(X_train,y_train, epochs=10)
```

```python
model.evaluate(X_test,y_test)
```
