```python
datagen = ImageDataGenerator(
    rotation_range=30,      # Rotate the image by 30 degree
    shear_range=0.2,        # distoration of .2
    zoom_range=0.2,         # zoom in 0r out by .2
    horizontal_flip=True,   # flip the image horizontaly
    width_shift_range=0.2,  # width shift of image by 0.2
    height_shift_range=0.2, # height shift of image by 0.2
    fill_mode= 'constant'   # this fill black on left-out pixel 
)
```

```python
img = image.img_to_array(img)   # to convert it into numpy array from PIL.Image

type(img)
```

```python
img.shape
```

```python
# to tell how many images need to regenerate

input_batch = img.reshape(1,200, 200, 3)  # only 1 is generated here
```

```python
# This will create 10 more images.

i = 0

for output in datagen.flow(input_batch, batch_size=1,save_to_dir='/Users/hirdeshpal/Jupiter Notebook/ML Algorithms/Deep Leaning/Datasets/aug'):
    i = i+1

    if i == 10:
        break
```
